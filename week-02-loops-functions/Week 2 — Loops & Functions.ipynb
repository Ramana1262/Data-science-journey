## for Loops & range() Function

A `for` loop is used to **repeat a block of code** a fixed number of times or go through each item in a sequence.

> 💡 **Real-life analogy:** Like reading a book page by page — you go through each page one by one until the book is finished.

## 1. Basic `for` Loop

**Syntax:**

for variable in sequence:
    # code to repeat


- `variable` → takes the value of each item one by one
- `sequence` → anything you can loop over (list, string, range...)

```python
# Example — loop through a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

# Output:
# apple
# banana
# cherry
```

## 2. Looping Through a String

A string is a sequence of characters — you can loop through each character.

```python
# Example
name = "Python"

for char in name:
    print(char)

# Output:
# P
# y
# t
# h
# o
# n
```
