**Introduction to Python**
Python is a high-level, interpreted, interactive, and object-oriented programming language created by Guido van Rossum in 1991. It is designed to be easy to read and write, making it a great choice for beginners and professionals alike.

Key Features of Python:

**High-Level Language**

 Python is a high-level language, meaning it abstracts away complex details (like memory management) and allows developers to focus on problem-solving rather than low-level operations.

**Interpreted Language**

Python is interpreted, meaning code is executed line by line (rather than compiled all at once). This makes debugging easier but can be slower than compiled languages like C++.

**Interactive Language**

Python supports an interactive mode, where you can write and execute code directly in the Python shell (REPL - Read-Eval-Print Loop), making it great for testing and learning.

**Object-Oriented (OOP) Language**

Python supports object-oriented programming (OOP), allowing developers to create classes and objects for structured and reusable code.

**Dynamically Typed**

Python does not require explicit variable type declarations (e.g., x = 10 instead of int x = 10), making it flexible but requiring careful type handling.

**Extensive Standard Library**

Python comes with a vast collection of built-in modules for tasks like file handling, web development, data science, and more.

**Why Python is Popular?**
Easy to learn and read (clean syntax).

Cross-platform (works on Windows, macOS, Linux).

Supports multiple programming paradigms (OOP, procedural, functional).

Widely used in web development (Django, Flask), data science (Pandas, NumPy), AI/ML (TensorFlow, PyTorch), automation, and more.

**Data Types and Type Conversion in Python**
**1. Data Types in Python**
Data types define the kind of value a variable can hold. Python has several built-in data types, categorized as:

A. Numeric Types
int (Integer): Whole numbers (e.g., 5, -10, 1000)

float (Floating-point): Decimal numbers (e.g., 3.14, -0.001, 2.0)

complex: Numbers with real and imaginary parts (e.g., 3 + 4j)

B. Text Type
str (String): Sequence of characters (e.g., "Hello", 'Python')

C. Boolean Type
bool: Represents True or False

D. Sequence Types
list: Ordered, mutable collection (e.g., [1, 2, 3])

tuple: Ordered, immutable collection (e.g., (1, 2, 3))

range: Sequence of numbers (e.g., range(5))

E. Mapping Type
dict (Dictionary): Key-value pairs (e.g., {"name": "Alice", "age": 25})

F. Set Types
set: Unordered, unique elements (e.g., {1, 2, 3})

frozenset: Immutable version of set

G. Binary Types
bytes: Immutable sequence of bytes (e.g., b"Hello")

bytearray: Mutable sequence of bytes

memoryview: Memory view of an object

**2. Type Conversion (Type Casting)**
Type conversion refers to changing a value from one data type to another.

A. Implicit Type Conversion (Automatic)
Python automatically converts smaller data types to larger ones to avoid data loss.

bool → int → float → complex

```python
#Example for Implicit type Conversion
a = 5      # int  
b = 2.5    # float  
result = a + b  # result becomes float (7.5)  
print(result)  

#Output: 7.5 (Python converts int to float automatically)
```

**B. Explicit Type Conversion (Manual)**
We manually convert data types using functions like:

int() → Converts to integer

float() → Converts to float

str() → Converts to string

list() → Converts to list

tuple() → Converts to tuple


```python
#Examples:

# String to Integer  
num_str = "10"  
num_int = int(num_str)  
print(num_int + 5)  # 15  

# Float to Integer  
pi = 3.14  
int_pi = int(pi)  # Truncates decimal part  
print(int_pi)     # 3  

# Integer to String  
age = 25  
age_str = str(age)  
print("Age: " + age_str)  # "Age: 25"  
```

Variables in Python are used to store data values. Unlike some other languages, Python does not require explicit declaration of variable types.

Must start with a letter or underscore (_).

Cannot start with a number.

Can only contain letters, numbers, and underscores.

Case-sensitive (age ≠ Age).

3. Local vs. Global Variables
Local Variable: Defined inside a function (accessible only within that function).

Global Variable: Defined outside a function (accessible everywhere).

```python
#Example of Local Variable

def my_func():  
    local_var = 5  # Local variable  
    print(local_var)  

my_func()  # Output: 5  
print(local_var)  # ❌ Error (not accessible outside)  
```

```python
#Example of Global Variable

global_var = 10  # Global variable  

def my_func():  
    print(global_var)  # Accessible  

my_func()  # Output: 10  
print(global_var)  # Output: 10  
```

4. Modifying a Global Variable Inside a Function
By default, if you try to modify a global variable inside a function, Python treats it as a new local variable unless you use the global keyword.

❌ Without global (Creates a Local Variable Instead)


```python
#Example
x = 10  # Global  

def modify():  
    x = 20  # New local variable (does not modify global x)  
    print("Inside function:", x)  

modify()  # Output: Inside function: 20  
print("Outside function:", x)  # Output: Outside function: 10 (unchanged) 
```

✅ Using global (Modifies the Actual Global Variable)

```python
x = 10  # Global  

def modify():  
    global x  # Declare x as global  
    x = 20    # Modifies the global x  
    print("Inside function:", x)  

modify()  # Output: Inside function: 20  
print("Outside function:", x)  # Output: Outside function: 20 (modified)  
```

5. Avoiding global (Alternative Approach: Using Return Values)
Instead of modifying a global variable directly, you can return the modified value and reassign it:


```python
x = 10  # Global  

def modify(num):  
    num = 20  # Local modification  
    return num  

x = modify(x)  # Reassign the returned value  
print(x)  # Output: 20
```

Key Takeaways
Global variables can be accessed anywhere.

To modify a global variable inside a function, use global.

Without global, Python creates a new local variable instead.

Best Practice: Avoid excessive use of global (use function parameters and return values instead).

Operators in Python
Operators are special symbols used to perform operations on variables and values. Python supports the following types of operators:

1. Arithmetic Operators
Used for mathematical calculations.


| Operator | Name            | Example   | Result |
|----------|-----------------|-----------|--------|
|   +      | Addition        | 5 + 3     | 8      |
| -        | Subtraction     | 10 - 2    | 8      |
| *        | Multiplication  | 4 * 3     | 12     |
| /        | Division        | 10 / 2    | 5.0    |
| %        | Modulus         | 10 % 3    | 3      |
| **       | Exponentiation  | 2 ** 3    | 8      |
| //       | Floor Division  | 10 // 3   | 3      |


```python
#example
print(5 + 3)   # 8  
print(10 // 3) # 3 (discards remainder)  
print(2 ** 4)  # 16 (2^4)  
```

2. Comparison (Relational) Operators
Compare two values and return True or False.


| Operator | Name                     | Example   | Result |
|----------|--------------------------|-----------|--------|
| ==       | Equal to                 | 5 == 5    | True   |
| !=       | Not equal to             | 5 != 3    | True   |
| >        | Greater than             | 10 > 5    | True   |
| <        | Less than                | 5 < 10    | True   |
| >=       | Greater than or equal    | 5 >= 5    | True   |
| <=       | Less than or equal       | 4 <= 5    | True   |


```python
x = 10  
print(x > 5)   # True  
print(x == 10) # True  
print(x != 20) # True  
```

3. Logical Operators
Combine conditional statements (and, or, not).

| Operator | Description                         | Example              | Result |
|----------|-------------------------------------|----------------------|--------|
| and      | True if both are true               | (5 > 3) and (2 < 4)  | True   |
| or       | True if at least one is true        | (5 > 3) or (2 > 4)   | True   |
| not      | Reverses the result (True→False)    | not(5 > 3)           | False  |

```python
age = 25  
print(age > 18 and age < 30)  # True  
print(age < 18 or age == 25)  # True  
print(not age == 25)          # False  
```

4. Assignment Operators
Assign values to variables.

| Operator | Example      | Equivalent to    |
|----------|--------------|------------------|
| =        | x = 5        | x = 5            |
| +=       | x += 3       | x = x + 3        |
| -=       | x -= 2       | x = x - 2        |
| *=       | x *= 4       | x = x * 4        |
| /=       | x /= 2       | x = x / 2        |
| %=       | x %= 3       | x = x % 3        |
| **=      | x **= 2      | x = x ** 2       |
| //=      | x //= 2      | x = x // 2       |

```python
x = 10  
x += 5  # x = 15  
x *= 2  # x = 30  
print(x)  # 30  
```

5. Membership Operators
Check if a value exists in a sequence (in, not in).

| Operator | Description                       | Example                | Result |
|----------|-----------------------------------|------------------------|--------|
| in       | True if value exists              | "a" in "apple"         | True   |
| not in   | True if value does not exist      | 10 not in [1, 2, 3]    | True   |

```python
fruits = ["apple", "banana", "cherry"]  
print("banana" in fruits)     # True  
print("mango" not in fruits)  # True  
```

6. Identity Operators
Compare memory locations of objects (is, is not).


| Operator | Description                    | Example       | Result              |
|----------|--------------------------------|---------------|---------------------|
| is       | True if same object            | x is y        | Depends             |
| is not   | True if not same object        | x is not y    | Depends             |

```python
a = [1, 2, 3]  
b = a  
c = [1, 2, 3]  

print(a is b)      # True (same object)  
print(a is c)      # False (different objects)  
print(a == c)      # True (same values)  
```
